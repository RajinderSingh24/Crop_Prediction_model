# Train Test Split

```python
col = ['Year', 'average_rain_fall_mm_per_year','pesticides_tonnes', 'avg_temp', 'Area', 'Item', 'hg/ha_yield']
df = df[col]
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0, shuffle=True)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Categorical to Numerical

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
ohe = OneHotEncoder(drop='first')
scale = StandardScaler()

preprocessor = ColumnTransformer(
        transformers = [
            ('StandardScale', scale, [0, 1, 2, 3]),
            ('OHE', ohe, [4, 5]),
        ],
        remainder='passthrough'
)
```

```python
preprocessor
```

```python
X_train_dummy = preprocessor.fit_transform(X_train)
X_test_dummy = preprocessor.transform(X_test)
```

```python
X_train_dummy
```

# Model Training

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,r2_score


models = {
    'lr':LinearRegression(),
    'lss':Lasso(),
    'Rid':Ridge(),
    'Dtr':DecisionTreeRegressor()
}
for name, md in models.items():
    md.fit(X_train_dummy,y_train)
    y_pred = md.predict(X_test_dummy)
    
    print(f"{name} : mae : {mean_absolute_error(y_test,y_pred)} score : {r2_score(y_test,y_pred)}")
```

```python
# Linear Regression Prediction and Visualization
lr_model = models['lr']
y_pred_lr = lr_model.predict(X_test_dummy)

# 1. Line Plot: Actual vs Predicted (Linear Regression)
plt.figure(figsize=(10, 6))
plt.plot(y_test.values[:50], label='Actual', marker='o')
plt.plot(y_pred_lr[:50], label='Predicted (Linear Regression)', marker='x')
plt.title('Actual vs Predicted Yield - Linear Regression (Top 50 Samples)')
plt.xlabel('Sample Index')
plt.ylabel('Crop Yield (hg/ha)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Scatter Plot: Predicted vs Actual (Linear Regression)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred_lr, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Actual Yield")
plt.ylabel("Predicted Yield")
plt.title("Linear Regression: Actual vs Predicted Yield Scatter Plot")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Model Selection

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train_dummy,y_train)
dtr.predict(X_test_dummy)
```

```python
# Visualize Actual vs Predicted for Decision Tree Regressor
y_pred = dtr.predict(X_test_dummy)

plt.figure(figsize=(10, 6))
plt.plot(y_test.values[:50], label='Actual', marker='o')
plt.plot(y_pred[:50], label='Predicted', marker='x')
plt.title('Actual vs Predicted Crop Yield (Top 50 Samples)')
plt.xlabel('Sample Index')
plt.ylabel('Crop Yield (hg/ha)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Predictive System

```python
def prediction(Year, average_rain_fall_mm_per_year, pesticides_tonnes, avg_temp, Area, Item):
    # Create an array of the input features
    features = np.array([[Year, average_rain_fall_mm_per_year, pesticides_tonnes, avg_temp, Area, Item]], dtype=object)

    # Transform the features using the preprocessor
    transformed_features = preprocessor.transform(features)

    # Make the prediction
    predicted_yield = dtr.predict(transformed_features).reshape(1, -1)

    return predicted_yield[0]

Year = 1990
average_rain_fall_mm_per_year =1485.0
pesticides_tonnes = 121.00
avg_temp = 16.37                   
Area = 'Albania'
Item = 'Maize'
result = prediction(Year, average_rain_fall_mm_per_year, pesticides_tonnes, avg_temp, Area, Item)
```

```python
result
```

```python
# Compare custom prediction with average yield for that crop
avg_yield_item = df[df['Item'] == Item]['hg/ha_yield'].mean()

plt.figure(figsize=(6, 4))
bars = plt.bar(['Predicted Yield', f'Avg {Item} Yield'], [result[0], avg_yield_item], color=['skyblue', 'lightgreen'])
plt.title(f"Custom Prediction vs Average {Item} Yield")
plt.ylabel("Yield (hg/ha)")
plt.grid(axis='y')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, f'{yval:.2f}', ha='center')

plt.tight_layout()
plt.show()
```
